# **Taller – Validez de una Función de Autocorrelación definida como suma de Gaussianas**

## 1. Planteamiento del problema

Se considera la siguiente función en tiempo continuo como modelo propuesto de autocorrelación:

$$
R(\tau)=\sum_{k\in{-1,0,1}} A_k
\exp\left(-\frac{(\tau-\mu_k)^2}{2\sigma_k^2}\right)
$$

donde los parámetros están definidos como:

$$
\mu_k = k
$$

$$
\sigma_k^2 =
\begin{cases}
\sigma^2, & k=0 \
\dfrac{\sigma^2}{|k|!}, & k\neq0
\end{cases}
$$

con las siguientes restricciones sobre las amplitudes:

$$
A_k>0
$$

$$
A_{-1}=A_1
$$

Esta función corresponde a una suma ponderada de funciones gaussianas centradas en distintos retardos temporales, y será analizada para determinar si puede ser considerada como una función de autocorrelación válida de un proceso estocástico estacionario en sentido amplio (WSS).


## 2. Propiedades Básicas de $R(\tau)$

Para que una función pueda ser considerada como una función de autocorrelación (ACF) válida de un proceso estacionario en sentido amplio (WSS), debe cumplir ciertas propiedades estructurales fundamentales. A continuación, se verifica si la función propuesta $R(\tau)$ satisface dichas condiciones.

---

### **Realidad**

La función está definida como:

$$
R(\tau)=\sum_{k\in{-1,0,1}} A_k
\exp\left(-\frac{(\tau-\mu_k)^2}{2\sigma_k^2}\right)
$$

Dado que:

* $A_k>0$
* $\mu_k \in \mathbb{R}$
* $\sigma_k^2>0$

cada término corresponde a una función gaussiana real. Por lo tanto, la suma de estos términos también es real:

$$
R(\tau)\in\mathbb{R}\quad \forall \tau
$$

---

### **Simetría**

Una ACF válida debe cumplir:

$$
R(\tau)=R(-\tau)
$$

Evaluando en $-\tau$:

$$
R(-\tau)=\sum_{k\in{-1,0,1}} A_k
\exp\left(-\frac{(-\tau-\mu_k)^2}{2\sigma_k^2}\right)
$$

Como:

$$
\mu_{-k}=-\mu_k
\quad \text{y} \quad
A_{-1}=A_1
$$

los términos para $k=1$ y $k=-1$ son simétricos respecto al origen, mientras que el término para $k=0$ es centrado en cero. Por lo tanto:

$$
R(-\tau)=R(\tau)
$$

---

### **Máximo en $\tau=0$**

Una función de autocorrelación debe cumplir:

$$
R(0)\ge R(\tau)\quad \forall \tau
$$

Evaluando en $\tau=0$:

$$
R(0)=\sum_{k\in{-1,0,1}} A_k
\exp\left(-\frac{\mu_k^2}{2\sigma_k^2}\right)
$$

Al aumentar $|\tau|$, el término $(\tau-\mu_k)^2$ crece, lo cual hace que cada gaussiana decrezca debido al exponente negativo. En consecuencia, la suma total también disminuye, cumpliéndose que:

$$
R(0)\ge R(\tau)
$$

---

Por lo tanto, la función $R(\tau)$ cumple con las propiedades básicas de realidad, simetría y máximo en el origen requeridas para ser una ACF candidata.


## 3 Condición de Positividad Definida

Una condición necesaria y suficiente para que una función $R(\tau)$ sea una función de autocorrelación (ACF) válida es que sea **positivamente definida**.

Formalmente, una función $R(\tau)$ es positivamente definida si para cualquier conjunto finito de escalares ${a_i}*{i=1}^{N}$ y retardos ${\tau_i}*{i=1}^{N}$ se cumple:

$$
\sum_{i=1}^{N}\sum_{j=1}^{N}
a_i a_j R(\tau_i-\tau_j)\ge0
$$

---

### Relación con matrices de Toeplitz

Esta propiedad es equivalente a exigir que la matriz construida a partir de muestras de $R(\tau)$:

$$
T_{ij}=R(\tau_i-\tau_j)
$$

sea **semidefinida positiva**, es decir, que todos sus autovalores sean mayores o iguales que cero:

$$
\lambda_i \ge 0
$$

En el caso de procesos estacionarios, esta matriz tiene estructura de tipo Toeplitz:

$$
T=
\begin{pmatrix}
R(0)        & R(\Delta)   & R(2\Delta) & R(3\Delta) & R(4\Delta) \\
R(\Delta)   & R(0)        & R(\Delta)  & R(2\Delta) & R(3\Delta) \\
R(2\Delta)  & R(\Delta)   & R(0)       & R(\Delta)  & R(2\Delta) \\
R(3\Delta)  & R(2\Delta)  & R(\Delta)  & R(0)       & R(\Delta) \\
R(4\Delta)  & R(3\Delta)  & R(2\Delta) & R(\Delta)  & R(0)
\end{pmatrix}
$$
---

Por lo tanto, si la matriz Toeplitz asociada a $R(\tau)$ presenta autovalores negativos, la función no es positivamente definida y no puede ser considerada como una ACF válida.


## 4. Teorema de Wiener–Khinchin

El teorema de Wiener–Khinchin establece que la densidad espectral de potencia $S(f)$ de un proceso estacionario en sentido amplio (WSS) está dada por la transformada de Fourier de su función de autocorrelación $R(\tau)$:

$$
S(f)=\mathcal{F}{R(\tau)}
$$

Una condición necesaria y suficiente para que $R(\tau)$ sea una función de autocorrelación válida es que su espectro asociado cumpla:

$$
S(f)\ge0 \quad \forall f
$$

---

En el caso particular de funciones gaussianas, se tiene que:

$$
\mathcal{F}\left\{
\exp\left(-\frac{(\tau-\mu)^2}{2\sigma^2}\right)
\right\}
=
\sqrt{2\pi}\,\sigma\;
\exp\left(-2\pi^2\sigma^2 f^2\right)\;
\exp\left(-j2\pi f \mu\right)
$$

Por lo tanto, el espectro asociado a $R(\tau)$ corresponde a una suma ponderada de gaussianas en frecuencia. La validez de la función como ACF dependerá de que dicha suma no genere valores negativos en $S(f)$.

---

En consecuencia, si el espectro presenta regiones donde $S(f)<0$, se viola la condición de Wiener–Khinchin y la función no puede ser considerada como una ACF válida.



```python
import torch
import numpy as np
from scipy.linalg import toeplitz
import matplotlib.pyplot as plt
import math  # para factorial

# Función para calcular el espectro
def compute_spectrum(R: torch.Tensor) -> torch.Tensor:
    R_shifted = torch.fft.fftshift(R)
    S = torch.fft.fft(R_shifted)
    S = torch.real(S)
    S = torch.fft.fftshift(S)
    return S

# Casos de prueba
casos = [
    {"A0": 1.0, "A1": 0.3, "sigma": 0.1},  # Cumple todas las condiciones
    {"A0": 1.0, "A1": 0.8, "sigma": 0.3},  # Puede fallar
    {"A0": 0.6, "A1": 0.5, "sigma": 0.7}   # No cumple
]

# Desplazamientos
mu = [-1, 0, 1]

# Tiempo discreto
N = 1024
t = torch.linspace(-3, 3, N)
freq = torch.linspace(-0.5, 0.5, N)

# ------------------------------------------------
# 1️⃣ Graficar las 3 funciones R(τ) juntas
plt.figure(figsize=(10,6))
for idx, caso in enumerate(casos, start=1):
    A0 = caso["A0"]
    A1 = caso["A1"]
    sigma = caso["sigma"]
    A = [A1, A0, A1]

    R = torch.zeros(N)
    for k in range(3):
        sigma_k = sigma if k==1 else sigma / math.sqrt(math.factorial(abs(k-1)))
        R += A[k] * torch.exp(-(t - mu[k])**2 / (2*sigma_k**2))
    plt.plot(t.numpy(), R.numpy(), label=f"Caso {idx}: A0={A0}, A1={A1}, σ={sigma}")

plt.title("Funciones R(τ) multi-gaussianas")
plt.xlabel("τ")
plt.ylabel("R(τ)")
plt.legend()
plt.grid()
plt.show()

# ------------------------------------------------
# 2️⃣ Gráfica de densidades espectrales (2 arriba, 1 abajo)
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(15,8))
gs = gridspec.GridSpec(2, 2, height_ratios=[1,1])  # 2 filas, 2 columnas

axes = []
axes.append(fig.add_subplot(gs[0,0]))  # Caso 1
axes.append(fig.add_subplot(gs[0,1]))  # Caso 2
axes.append(fig.add_subplot(gs[1,:]))  # Caso 3 (abajo ocupando 2 columnas)

for idx, caso in enumerate(casos):
    A0 = caso["A0"]
    A1 = caso["A1"]
    sigma = caso["sigma"]
    A = [A1, A0, A1]

    R = torch.zeros(N)
    for k in range(3):
        sigma_k = sigma if k==1 else sigma / math.sqrt(math.factorial(abs(k-1)))
        R += A[k] * torch.exp(-(t - mu[k])**2 / (2*sigma_k**2))

    # Error de simetría
    error_sim = torch.max(torch.abs(R - torch.flip(R, dims=[0]))).item()

    # Autovalor mínimo Toeplitz
    R_np = R.numpy()
    T = toeplitz(R_np)
    eig_min = np.min(np.linalg.eigvalsh(T))

    # Espectro y valor mínimo S(f)
    S = compute_spectrum(R)
    min_S = torch.min(S).item()

    # Graficar densidad espectral
    axes[idx].plot(freq.numpy(), S.numpy())
    axes[idx].set_title(f"Caso {idx+1}")
    axes[idx].set_xlabel("Frecuencia")
    axes[idx].set_ylabel("S(f)")
    axes[idx].grid(True)

    # Imprimir resultados
    print(f"Caso {idx+1}:")
    print(f"Error de simetría = {error_sim:.2e}")
    print(f"Autovalor mínimo Toeplitz = {eig_min:.4f}")
    print(f"Valor mínimo S(f) = {min_S:.2e}\n")

plt.suptitle("Densidades espectrales S(f) para los 3 casos")
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

---

## 2. Interpretación de cada métrica

### a) Error de simetría: $|R(t)-R(-t)|$

* Para los tres casos, el error de simetría es prácticamente **0**, indicando que las señales son **muy simétricas respecto a $t=0$**.
* Esto es consistente con la forma de la función:

$
R(\tau) = \sum_{k=-1}^{1} A_k \exp\Big[-\frac{(\tau-\mu_k)^2}{2\sigma_k^2}\Big]
$

que es **par** por construcción $A_{-1}=A_1$ y $\mu_{-1}=-1, \mu_1=1$.

* Implicación: cualquier análisis de autocorrelación, espectro o Toeplitz **no se ve afectado por asimetrías**, ya que no existen.

---

### b) Autovalor mínimo de la matriz Toeplitz

* La matriz Toeplitz se construye a partir de los valores de $R(t)$ y sus autovalores indican la **positividad y estabilidad**:

  * Caso 1: -55.37
  * Caso 2: -220.10
  * Caso 3: -252.66

* Observaciones:

  * Todos los casos tienen **autovalores negativos**, por lo que las matrices **no son definidas positivas**.
  * La magnitud del autovalor negativo crece con $A_0$ y $\sigma$, es decir, señales más altas y más anchas producen valores negativos mayores.
  * Esto puede ser relevante si se usan estas matrices para **procesos estocásticos o filtros**, porque matrices no definidas positivas pueden causar problemas numéricos.

---

### c) Valor mínimo del espectro $S(f)$

* Valores obtenidos:

  * Caso 1: -1.93e-06 → prácticamente 0
  * Caso 2: -49.4 → negativo significativo
  * Caso 3: -10.3 → negativo

* Interpretación:

  * Caso 1: el mínimo negativo es **solo efecto numérico** por la FFT discreta y se puede considerar positivo.
  * Casos 2 y 3: los mínimos son significativos, indicando que las gaussianas más anchas generan **oscilaciones importantes en frecuencia**.
  * Esto se debe a que un σ grande dentro de un dominio limitado provoca **efecto de borde (Gibbs)** en la FFT.

* Implicación: Para análisis espectral, **solo el caso 1 cumple estrictamente la condición de espectro positivo**.

---

## 3. Relación entre los parámetros y los resultados

| Parámetro | Observación                                                                                                      |
| --------- | ---------------------------------------------------------------------------------------------------------------- |
| $A_0$     | Al aumentar $A_0$, el autovalor mínimo de Toeplitz y los valores negativos del espectro aumentan en magnitud.    |
| $\sigma$  | σ pequeño → espectro suave y mínimo cercano a 0. σ grande → espectro con valores negativos fuertes y artefactos. |
| $A_1$     | El offset $A_1$ desplaza la señal, pero no afecta significativamente el mínimo del espectro.                     |

---

## 4. Conclusión general

1. **Simetría:** todas las señales son simétricas → error de simetría ≈ 0.
2. **Autovalor mínimo Toeplitz:** todos negativos → matrices no definidas positivas, creciente con $A_0$ y σ.
3. **Espectro $S(f)$:**

   * Caso 1 → mínimo ≈ 0 → cumple condición de espectro positivo.
   * Casos 2 y 3 → mínimos negativos → no cumplen.

**Interpretación física:**

* Una gaussiana estrecha (Caso 1) produce un espectro suave y estable.
* Una gaussiana más ancha (Casos 2 y 3) dentro de un dominio finito genera artefactos en la FFT y autovalores negativos → precaución si se usa para modelar energía positiva o filtros.
